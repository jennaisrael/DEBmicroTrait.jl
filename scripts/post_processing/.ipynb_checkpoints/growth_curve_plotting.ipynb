## Plot the time series variables over the growth curve

```python
df.columns[4]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(7, 7))
axs = axs.flat

for i in np.arange(1,5):
    axs[i-1].plot(df.time,df.iloc[:,[i]],'-o')
    axs[i-1].set_xlabel('Time')
    axs[i-1].set_ylabel(df.columns[i])
    axs[i-1].set_xlim(0,50)
   
fig.suptitle(strain + " in 1/10 R2A")
plt.show()
```
